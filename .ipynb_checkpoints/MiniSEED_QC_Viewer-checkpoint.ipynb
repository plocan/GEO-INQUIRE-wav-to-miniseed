# MiniSEED Quality Control & PPSD Viewer
**Verify your processed files**

This notebook helps you:
1. **Validate StationXML** - Check schema compliance (IRIS validator)
2. **Read MiniSEED files** - Verify they load correctly
3. **View waveforms** - See the signal data
4. **Check StationXML response** - View sensitivity and frequency response
5. **Compute PPSD** - Probabilistic Power Spectral Density analysis

## Requirements
- Java 8 or 11 (for StationXML validator)
- ObsPy (already installed if you ran the main tool)
- matplotlib for plotting

## Workflow
This notebook is intended for checking a small subset of files (a few examples), not the full dataset.
Run cells in order. In Cell 4 you can select the MiniSEED files you want; the StationXML will be **automatically matched** from the same folder.

```python
# Cell 3: Import libraries
import os
from tkinter import Tk, filedialog
import obspy
from obspy import read, read_inventory
from obspy.signal import PPSD
import matplotlib.pyplot as plt
import numpy as np

# Enable interactive plots in notebook
%matplotlib inline
plt.rcParams['figure.figsize'] = [14, 8]
plt.rcParams['figure.dpi'] = 100

print(f"ObsPy version: {obspy.__version__}")
print("✓ Libraries loaded successfully")
```

```python
# Cell 4: Select MiniSEED files (multi-select)
# Selected files will be used automatically in later cells.

import os
from tkinter import Tk, filedialog

qc_folder = None
qc_mseed_files = []
qc_xml_files = []
qc_pairs = []

root = Tk()
root.withdraw()
root.attributes('-topmost', True)
files = filedialog.askopenfilenames(
    title="Select MiniSEED files",
    filetypes=[("MiniSEED files", "*.mseed"), ("All files", "*.*")]
)
root.destroy()

if files:
    qc_mseed_files = list(files)
    qc_folder = os.path.dirname(qc_mseed_files[0])
    
    # Collect StationXML files from the same folder
    qc_xml_files = [
        os.path.join(qc_folder, f)
        for f in os.listdir(qc_folder)
        if f.lower().endswith(".station.xml")
    ]
    
    xml_map = {os.path.basename(x).replace(".station.xml", ""): x for x in qc_xml_files}
    for m in qc_mseed_files:
        base = os.path.splitext(os.path.basename(m))[0]
        qc_pairs.append((m, xml_map.get(base)))

    print(f"Selected MiniSEED files: {len(qc_mseed_files)}")
    print(f"StationXML files in folder: {len(qc_xml_files)}")

    missing_xml = [os.path.basename(m) for m, x in qc_pairs if not x]
    if missing_xml:
        print("\nMiniSEED files without matching StationXML:")
        for name in missing_xml:
            print(f"  - {name}")
else:
    print("No files selected. You can still pick files manually later.")

# Optional settings for batch plots (Cells 8-10)
# Set qc_max_files to an integer to limit plotting, or None for all.
qc_max_files = None
# Set qc_show_ppsd_frequency to False to skip frequency-axis PPSD plot.
qc_show_ppsd_frequency = True
# Set qc_special_handling to None or a valid PPSD special_handling string
# (e.g., "hydrophone", "ringlaser", "infrasound").
qc_special_handling = "hydrophone"
# Set qc_response_output to control response plot units
# (e.g., "VEL", "DISP", "ACC").
qc_response_output = "VEL"
```

```python
# Cell 5: Download IRIS StationXML Validator (run once)
import urllib.request
import os

DOWNLOAD_URLS = [
    ("Maven Central", "https://repo1.maven.org/maven2/edu/iris/dmc/stationxml-validator/1.7.3/stationxml-validator-1.7.3-jar-with-dependencies.jar"),
]
VALIDATOR_JAR = "stationxml-validator-1.7.3.jar"

if os.path.exists(VALIDATOR_JAR):
    print(f"✓ Validator already downloaded: {VALIDATOR_JAR}")
else:
    for name, url in DOWNLOAD_URLS:
        print(f"Downloading from {name}...")
        try:
            urllib.request.urlretrieve(url, VALIDATOR_JAR)
            print(f"✓ Downloaded successfully!")
            break
        except Exception as e:
            print(f"✗ Failed: {e}")
            print(f"\nManual download: https://search.maven.org/artifact/edu.iris.dmc/stationxml-validator")
            print(f"Save the JAR file as '{VALIDATOR_JAR}' in this folder.")
```

```python
# Cell 6: Validate StationXML files with IRIS Validator
import subprocess
import os
from tkinter import Tk, filedialog

VALIDATOR_JAR = "stationxml-validator-1.7.3.jar"

def validate_stationxml(file_path):
    """Validate a single StationXML file."""
    print(f"\n{'='*60}")
    print(f"Validating: {os.path.basename(file_path)}")
    print('='*60)
    
    try:
        result = subprocess.run(
            ['java', '-jar', VALIDATOR_JAR, '--input', file_path],
            capture_output=True, text=True, timeout=60
        )
        if result.stdout:
            print(result.stdout)
        if result.stderr:
            print(result.stderr)
        
        if result.returncode == 0:
            print("\n✓ VALIDATION PASSED")
            return True
        else:
            print("\n✗ VALIDATION FAILED")
            return False
    except Exception as e:
        print(f"✗ Error: {e}")
        return False

# Check validator exists
if not os.path.exists(VALIDATOR_JAR):
    print("Validator not found. Run Cell 5 first to download it.")
else:
    # Use folder selection if available
    if 'qc_xml_files' in globals() and qc_xml_files:
        files = qc_xml_files
        print(f"Using {len(files)} StationXML file(s) from selected folder.\n")
    else:
        root = Tk()
        root.withdraw()
        root.attributes('-topmost', True)
        
        files = filedialog.askopenfilenames(
            title="Select StationXML files to validate",
            filetypes=[("XML files", "*.xml"), ("All files", "*.*")]
        )
        root.destroy()
    
    if files:
        print(f"Selected {len(files)} file(s)\n")
        passed, failed = 0, 0
        for f in files:
            if validate_stationxml(f):
                passed += 1
            else:
                failed += 1
        
        print(f"\n{'='*60}")
        print(f"SUMMARY: {passed} passed, {failed} failed out of {len(files)}")
        print('='*60)
    else:
        print("No files selected.")
```

```python
# Cell 7: Read MiniSEED file and AUTO-LOAD matching StationXML
# The StationXML file must have the same base name (e.g., file.mseed -> file.station.xml)

from obspy import read, read_inventory

mseed_file = None
xml_file = None

# Use selected files from Cell 4
if 'qc_mseed_files' in globals() and qc_mseed_files:
    mseed_file = qc_mseed_files[0]
    print("Using first selected MiniSEED file for plots.")
    print("For batch QC across all selected files, run Cell 13.")
    
    # Get matching XML from qc_pairs if available
    if 'qc_pairs' in globals() and qc_pairs:
        for m, x in qc_pairs:
            if m == mseed_file:
                xml_file = x
                break
else:
    print("No files selected. Run Cell 4 first.")

if mseed_file:
    print(f"MiniSEED: {os.path.basename(mseed_file)}")
    print("="*60)
    
    # Read the MiniSEED file
    st = read(mseed_file)
    
    # AUTO-FIND matching StationXML
    if not xml_file:
        base_name = os.path.splitext(mseed_file)[0]
        xml_file = base_name + ".station.xml"
    
    inv = None
    if xml_file and os.path.exists(xml_file):
        print(f"StationXML: {os.path.basename(xml_file)} (auto-matched)")
        inv = read_inventory(xml_file)
    else:
        print(f"StationXML not found for {os.path.basename(mseed_file)}")
        print("Expected filename pattern: <name>.mseed -> <name>.station.xml")
    
    print("\n" + "="*60)
    print("MINISEED FILE INFO")
    print("="*60)
    print(st)
    
    for tr in st:
        print(f"\nTrace: {tr.id}")
        print(f"   Network:     {tr.stats.network}")
        print(f"   Station:     {tr.stats.station}")
        print(f"   Location:    {tr.stats.location}")
        print(f"   Channel:     {tr.stats.channel}")
        print(f"   Start time:  {tr.stats.starttime}")
        print(f"   End time:    {tr.stats.endtime}")
        print(f"   Duration:    {tr.stats.endtime - tr.stats.starttime:.2f} seconds")
        print(f"   Sample rate: {tr.stats.sampling_rate} Hz")
        print(f"   Samples:     {tr.stats.npts}")
        print(f"   Data range:  {tr.data.min()} to {tr.data.max()} counts")
    
    # Show StationXML info if loaded
    if inv:
        print("\n" + "="*60)
        print("STATIONXML RESPONSE INFO")
        print("="*60)
        for network in inv:
            for station in network:
                for channel in station:
                    if channel.response and channel.response.instrument_sensitivity:
                        sens = channel.response.instrument_sensitivity
                        print(f"\n   Sensitivity:")
                        print(f"      Value: {sens.value:.6f}")
                        print(f"      Units: {sens.input_units} -> {sens.output_units}")
                        print(f"      Reference frequency: {sens.frequency} Hz")
    
    print("\nFiles loaded.")
```

```python
# Cell 8: Plot the waveform

from obspy import read
import os

def _plot_waveform_for_stream(stream, label):
    print(f"WAVEFORM PLOT: {label}")
    print("="*60)
    
    fig = plt.figure(figsize=(14, 6))
    
    for i, tr in enumerate(stream):
        ax = fig.add_subplot(len(stream), 1, i + 1)
        
        # Create time axis
        times = np.arange(tr.stats.npts) / tr.stats.sampling_rate
        
        ax.plot(times, tr.data, 'b-', linewidth=0.5)
        ax.set_xlabel('Time (seconds)')
        ax.set_ylabel('Counts')
        ax.set_title(f"{tr.id} - {tr.stats.starttime.strftime('%Y-%m-%d %H:%M:%S')} UTC")
        ax.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()

if 'qc_mseed_files' in globals() and qc_mseed_files:
    files_to_plot = qc_mseed_files
    if 'qc_max_files' in globals() and qc_max_files is not None:
        files_to_plot = qc_mseed_files[:qc_max_files]
    for mseed_file in files_to_plot:
        try:
            st_local = read(mseed_file)
            _plot_waveform_for_stream(st_local, os.path.basename(mseed_file))
        except Exception as e:
            print(f"Error plotting waveform for {os.path.basename(mseed_file)}: {e}")
else:
    if 'st' in dir() and st:
        _plot_waveform_for_stream(st, "current selection")
    else:
        print("No MiniSEED data loaded. Run Cell 7 first.")
```

```python
# Cell 9: Plot the instrument response (amplitude and phase)

from obspy import read_inventory
import os

def _find_xml_for_mseed(mseed_file):
    if 'qc_pairs' in globals() and qc_pairs:
        for m, x in qc_pairs:
            if m == mseed_file:
                return x
    base_name = os.path.splitext(mseed_file)[0]
    return base_name + ".station.xml"

def _response_units_label(output_type):
    output_type = (output_type or "VEL").upper()
    if output_type == "DISP":
        return "m"
    if output_type == "ACC":
        return "m/s²"
    return "m/s"


def _plot_response_for_inventory(inv_local, label):
    print(f"INSTRUMENT RESPONSE PLOT: {label}")
    print("="*60)
    
    response_output = (
        qc_response_output
        if 'qc_response_output' in globals() and qc_response_output
        else "VEL"
    )
    units_label = _response_units_label(response_output)
    
    for network in inv_local:
        for station in network:
            for channel in station:
                if channel.response:
                    print(f"\nPlotting response for {network.code}.{station.code}.{channel.location_code}.{channel.code}")
                    
                    try:
                        freqs = np.logspace(-2, np.log10(channel.sample_rate / 2), 500)
                        resp_complex = channel.response.get_evalresp_response_for_frequencies(
                            freqs, output=response_output
                        )
                        
                        fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))
                        
                        ax1.loglog(freqs, np.abs(resp_complex), 'b-', linewidth=2)
                        ax1.set_xlabel('Frequency (Hz)')
                        ax1.set_ylabel(f"Amplitude (counts/{units_label})")
                        ax1.set_title('Amplitude Response')
                        ax1.grid(True, alpha=0.3, which='both')
                        ax1.axvline(x=channel.sample_rate / 2, color='r', linestyle='--',
                                   label=f'Nyquist ({channel.sample_rate / 2} Hz)')
                        ax1.legend()
                        
                        phase_deg = np.angle(resp_complex, deg=True)
                        ax2.semilogx(freqs, phase_deg, 'g-', linewidth=2)
                        ax2.set_xlabel('Frequency (Hz)')
                        ax2.set_ylabel('Phase (degrees)')
                        ax2.set_title('Phase Response')
                        ax2.grid(True, alpha=0.3, which='both')
                        ax2.set_ylim([-180, 180])
                        
                        plt.suptitle(
                            f"Instrument Response: {label} | {network.code}.{station.code}.{channel.code}"
                        )
                        plt.tight_layout()
                        plt.show()
                        
                    except Exception as e:
                        print(f"Error plotting response: {e}")

if 'qc_mseed_files' in globals() and qc_mseed_files:
    files_to_plot = qc_mseed_files
    if 'qc_max_files' in globals() and qc_max_files is not None:
        files_to_plot = qc_mseed_files[:qc_max_files]
    for mseed_file in files_to_plot:
        xml_file = _find_xml_for_mseed(mseed_file)
        if not xml_file or not os.path.exists(xml_file):
            print(f"No StationXML found for {os.path.basename(mseed_file)}")
            continue
        try:
            inv_local = read_inventory(xml_file)
            _plot_response_for_inventory(inv_local, os.path.basename(mseed_file))
        except Exception as e:
            print(f"Error reading StationXML for {os.path.basename(mseed_file)}: {e}")
else:
    if 'inv' in dir() and inv:
        _plot_response_for_inventory(inv, "current selection")
    else:
        print("No StationXML loaded. Run Cell 7 first.")
```

```python
# Cell 10: Compute and plot PPSD

from obspy import read, read_inventory
import os

def _find_xml_for_mseed(mseed_file):
    if 'qc_pairs' in globals() and qc_pairs:
        for m, x in qc_pairs:
            if m == mseed_file:
                return x
    base_name = os.path.splitext(mseed_file)[0]
    return base_name + ".station.xml"

def _get_input_units(inv_local):
    try:
        channel = inv_local[0][0][0]
        sens = channel.response.instrument_sensitivity if channel.response else None
        if sens and sens.input_units:
            return sens.input_units
    except Exception:
        pass
    return "unknown"

def _run_ppsd_for_stream(stream, inv_local, label):
    print("COMPUTING PPSD")
    print(f"File: {label}")
    print("="*60)
    
    tr = stream[0]
    input_units = _get_input_units(inv_local)
    display_units = (
        qc_units_label
        if 'qc_units_label' in globals() and qc_units_label
        else input_units
    )
    power_label = f"Power Spectral Density (dB re 1 {display_units}²/Hz)"
    asd_label = f"ASD (dB re 1 {display_units}/√Hz)"
    title_units = f" ({display_units})" if display_units else ""
    
    try:
        _special_handling = (
            qc_special_handling
            if 'qc_special_handling' in globals() and qc_special_handling
            else None
        )
        ppsd = PPSD(
            tr.stats,
            inv_local,
            ppsd_length=60.0,
            overlap=0.5,
            special_handling=_special_handling,
            db_bins=(-200, -20, 1.0),
            period_limits=(0.01, 100)
        )
    
        ppsd.add(stream)
    
        print(f"PPSD computed for: {tr.id}")
        print(f"Number of segments: {len(ppsd.times_processed)}")
        
        fig = ppsd.plot(show=False)
        if fig and fig.axes:
            fig.axes[0].set_ylabel(power_label)
        fig.suptitle(f"PPSD (power): {label}{title_units}", y=1.02)
        plt.show()
        
        if 'qc_show_ppsd_frequency' not in globals() or qc_show_ppsd_frequency:
            fig = ppsd.plot(xaxis_frequency=True, show=False)
            if fig and fig.axes:
                fig.axes[0].set_ylabel(power_label)
                fig.axes[0].set_xlabel("Frequency (Hz)")
            fig.suptitle(f"PPSD (power, frequency): {label}{title_units}", y=1.02)
            plt.show()
        
        print("\n" + "="*60)
        print("PPSD STATISTICS")
        print("="*60)
        
        (periods, mean) = ppsd.get_mean()
        (periods, p50) = ppsd.get_percentile(50)
        (periods, p5) = ppsd.get_percentile(5)
        (periods, p95) = ppsd.get_percentile(95)
        
        freqs = 1.0 / periods
        key_freqs = [1, 10, 50, 100]
        print("\nPower levels at key frequencies:")
        print(f"{'Freq (Hz)':<12} {'Mean (dB)':<12} {'Median (dB)':<12} {'5%-95% range'}")
        print("-" * 60)
        
        for kf in key_freqs:
            if kf <= freqs.max() and kf >= freqs.min():
                idx = np.argmin(np.abs(freqs - kf))
                print(f"{kf:<12} {mean[idx]:<12.1f} {p50[idx]:<12.1f} [{p5[idx]:.1f} to {p95[idx]:.1f}]")
        
        print("\nOverall statistics:")
        print(f"Mean power (all freq): {np.mean(mean):.1f} dB")
        print(f"Min power:  {np.min(mean):.1f} dB at {1/periods[np.argmin(mean)]:.2f} Hz")
        print(f"Max power:  {np.max(mean):.1f} dB at {1/periods[np.argmax(mean)]:.2f} Hz")
        
        # ASD output (Pa/√Hz) from PPSD power (Pa²/Hz)
        asd_mean = mean / 2.0
        print("\nASD (amplitude spectral density) at key frequencies:")
        print(f"{'Freq (Hz)':<12} {f'Mean ASD (dB re 1 {display_units}/√Hz)':<30}")
        print("-" * 45)
        for kf in key_freqs:
            if kf <= freqs.max() and kf >= freqs.min():
                idx = np.argmin(np.abs(freqs - kf))
                print(f"{kf:<12} {asd_mean[idx]:<30.1f}")
        
        plt.figure(figsize=(12, 6))
        plt.semilogx(freqs, asd_mean, 'b-')
        plt.xlabel('Frequency (Hz)')
        plt.ylabel(asd_label)
        plt.title(f"ASD (mean): {label}{title_units}")
        plt.grid(True, alpha=0.3, which='both')
        plt.show()
        
    except Exception as e:
        print(f"Error computing PPSD: {e}")
        import traceback
        traceback.print_exc()
        
        from scipy import signal
        f, Pxx = signal.welch(tr.data, tr.stats.sampling_rate, 
                             nperseg=min(4096, len(tr.data)//4))
        
        plt.figure(figsize=(12, 6))
        plt.semilogy(f, Pxx)
        plt.xlabel('Frequency (Hz)')
        plt.ylabel('Power Spectral Density (counts²/Hz)')
        plt.title(
            f"PSD (uncalibrated): {tr.id}"
            f"{(' (' + qc_units_label + ')') if 'qc_units_label' in globals() and qc_units_label else ''}"
        )
        plt.grid(True, alpha=0.3)
        plt.xlim([0, tr.stats.sampling_rate/2])
        plt.show()

if 'qc_mseed_files' in globals() and qc_mseed_files:
    files_to_plot = qc_mseed_files
    if 'qc_max_files' in globals() and qc_max_files is not None:
        files_to_plot = qc_mseed_files[:qc_max_files]

    units_seen = set()
    for mseed_file in files_to_plot:
        xml_file = _find_xml_for_mseed(mseed_file)
        if not xml_file or not os.path.exists(xml_file):
            continue
        try:
            inv_local = read_inventory(xml_file)
            units_seen.add(_get_input_units(inv_local))
        except Exception:
            pass
    units_seen.discard("unknown")
    if len(units_seen) > 1:
        qc_units_label = "mixed units"
    elif len(units_seen) == 1:
        qc_units_label = next(iter(units_seen))
    else:
        qc_units_label = "unknown"

    for mseed_file in files_to_plot:
        xml_file = _find_xml_for_mseed(mseed_file)
        if not xml_file or not os.path.exists(xml_file):
            print(f"No StationXML found for {os.path.basename(mseed_file)}")
            continue
        try:
            st_local = read(mseed_file)
            inv_local = read_inventory(xml_file)
            _run_ppsd_for_stream(st_local, inv_local, os.path.basename(mseed_file))
        except Exception as e:
            print(f"Error processing {os.path.basename(mseed_file)}: {e}")
else:
    if 'st' in dir() and st and 'inv' in dir() and inv:
        _run_ppsd_for_stream(st, inv, "current selection")
    else:
        print("Required data not loaded.")
        if 'st' not in dir() or not st:
            print("- Run Cell 7 to load MiniSEED file")
        if 'inv' not in dir() or not inv:
            print("- StationXML not found or not loaded")
```

```python
# Cell 11: Response chain diagnostic
# Run after Cell 7 (MiniSEED + StationXML loaded)

import numpy as np

if 'st' in dir() and st and 'inv' in dir() and inv:
    tr = st[0]
    print("="*60)
    print("RESPONSE CHAIN DIAGNOSTIC")
    print("="*60)
    
    for network in inv:
        for station in network:
            for channel in station:
                if not channel.response:
                    continue
                resp = channel.response
                print(f"\nChannel: {network.code}.{station.code}.{channel.location_code}.{channel.code}")
                
                if resp.instrument_sensitivity:
                    sens = resp.instrument_sensitivity
                    print(f"Instrument sensitivity: {sens.value:.6f} {sens.output_units}/{sens.input_units}")
                    print(f"Reference frequency: {sens.frequency} Hz")
                else:
                    print("Instrument sensitivity: NOT FOUND")
                    sens = None
                
                print("\nResponse stages:")
                total_gain = 1.0
                for stage in resp.response_stages:
                    stage_gain = stage.stage_gain if stage.stage_gain is not None else 1.0
                    total_gain *= stage_gain
                    print(
                        f"  Stage {stage.stage_sequence_number}: {type(stage).__name__} | "
                        f"Gain={stage_gain} at {stage.stage_gain_frequency} Hz | "
                        f"Units={stage.input_units}->{stage.output_units}"
                    )
                
                print(f"\nProduct of stage gains: {total_gain:.6f}")
                if sens is not None:
                    print(f"Instrument sensitivity (reported): {sens.value:.6f}")
                    ratio = sens.value / total_gain if total_gain != 0 else np.nan
                    print(f"Ratio (sensitivity / stage_gain_product): {ratio:.6f}")
                
                counts_min = float(np.min(tr.data))
                counts_max = float(np.max(tr.data))
                counts_peak = max(abs(counts_min), abs(counts_max))
                print(f"\nData range: {counts_min:.1f} to {counts_max:.1f} counts")
                print(f"Peak amplitude: {counts_peak:.1f} counts")
                
                if sens is not None and sens.value > 0:
                    pa_peak = counts_peak / sens.value
                    print(f"Estimated peak pressure: {pa_peak:.6e} Pa")
                    print(f"Estimated peak level: {20*np.log10(pa_peak/1e-6):.1f} dB re 1 µPa")
else:
    print("Required data not loaded.")
    print("- Run Cell 7 first to load MiniSEED and StationXML")
```

```python
# Cell 12: Compare WAV vs MiniSEED amplitude (same base name)
# Requires Cell 7 to load MiniSEED

import numpy as np
import os
import soundfile as sf
from obspy import read

# Settings for WAV float interpretation (match your metadata)
wav_float_units = "normalized"  # normalized | volts | counts
nbits = 24
ref_voltage = 2.5

if 'mseed_file' in dir() and mseed_file:
    # Auto-match WAV by base name
    base = os.path.splitext(mseed_file)[0]
    wav_file = base + ".wav"
    
    if not os.path.exists(wav_file):
        print(f"WAV not found with same name: {os.path.basename(wav_file)}")
        print("Please check the WAV filename or rename to match the MiniSEED.")
    else:
        print(f"MiniSEED: {os.path.basename(mseed_file)}")
        print(f"WAV:     {os.path.basename(wav_file)}")
        print("="*60)
        
        # Read WAV
        info = sf.info(wav_file)
        wav_data, wav_sr = sf.read(wav_file, always_2d=False)
        wav_data = np.asarray(wav_data)
        
        # Read MiniSEED data (reload if needed)
        if 'st' not in globals() or not st:
            st = read(mseed_file)
        tr = st[0]
        mseed_data = np.asarray(tr.data)
        
        # Convert to float for safe math
        wav_f = wav_data.astype(np.float64)
        mseed_f = mseed_data.astype(np.float64)
        
        # Optional: scale WAV float to counts for comparison
        wav_counts = None
        if info.subtype in ("FLOAT", "DOUBLE"):
            if wav_float_units == "normalized":
                wav_counts = np.round(wav_f * (2 ** (nbits - 1) - 1)).astype(np.float64)
            elif wav_float_units == "volts":
                counts_per_v = (2 ** (nbits - 1)) / ref_voltage
                wav_counts = np.round(wav_f * counts_per_v).astype(np.float64)
            elif wav_float_units == "counts":
                wav_counts = wav_f.copy()
        
        # Basic stats (mean + RMS, raw and demeaned)
        def stats(x):
            mean = float(np.mean(x))
            rms = float(np.sqrt(np.mean(x**2)))
            rms_demean = float(np.sqrt(np.mean((x - mean)**2)))
            return {
                "min": float(np.min(x)),
                "max": float(np.max(x)),
                "mean": mean,
                "rms": rms,
                "rms_demean": rms_demean,
            }
        
        wav_stats = stats(wav_f)
        mseed_stats = stats(mseed_f)
        
        print("WAV stats:")
        print(f"  subtype: {info.subtype}")
        print(f"  min/max: {wav_stats['min']} / {wav_stats['max']}")
        print(f"  mean: {wav_stats['mean']:.6f}")
        print(f"  rms:  {wav_stats['rms']:.6f}")
        print(f"  rms (demeaned): {wav_stats['rms_demean']:.6f}")
        print(f"  sample rate: {wav_sr} Hz")
        
        print("\nMiniSEED stats:")
        print(f"  dtype: {mseed_data.dtype}")
        print(f"  min/max: {mseed_stats['min']} / {mseed_stats['max']}")
        print(f"  mean: {mseed_stats['mean']:.6f}")
        print(f"  rms:  {mseed_stats['rms']:.6f}")
        print(f"  rms (demeaned): {mseed_stats['rms_demean']:.6f}")
        print(f"  sample rate: {tr.stats.sampling_rate} Hz")
        
        if wav_counts is not None:
            wavc_stats = stats(wav_counts)
            print("\nWAV (scaled to counts) stats:")
            print(f"  min/max: {wavc_stats['min']} / {wavc_stats['max']}")
            print(f"  mean: {wavc_stats['mean']:.6f}")
            print(f"  rms (demeaned): {wavc_stats['rms_demean']:.6f}")
            
            # Compare RMS in counts (demeaned)
            if wavc_stats['rms_demean'] > 0:
                ratio = mseed_stats['rms_demean'] / wavc_stats['rms_demean']
                print("\nAmplitude ratio (MiniSEED RMS_demean / WAV_counts RMS_demean):")
                print(f"  {ratio:.6e}")
        

else:
    print("MiniSEED not loaded. Run Cell 7 first.")
```

```python
# Cell 13: Run QC on folder and show summary plot
# Uses the files selected in Cell 4 (qc_pairs)

import os
import numpy as np
from obspy import read, read_inventory
from obspy.signal import PPSD

if 'qc_pairs' in globals() and qc_pairs:
    print("QC SUMMARY")
    print("=" * 60)
    
    summary_rows = []
    combined_ppsd = None
    summary_units = None
    units_seen = set()
    _special_handling = (
        qc_special_handling
        if 'qc_special_handling' in globals() and qc_special_handling
        else None
    )
    
    def _get_input_units_from_inv(inv_local):
        try:
            channel = inv_local[0][0][0]
            sens = channel.response.instrument_sensitivity if channel.response else None
            if sens and sens.input_units:
                return sens.input_units
        except Exception:
            pass
        return "unknown"
    
    for mseed_path, xml_path in qc_pairs:
        base = os.path.basename(mseed_path)
        xml_ok = bool(xml_path and os.path.exists(xml_path))
        
        # Read MiniSEED
        st = read(mseed_path)
        tr = st[0]
        
        # Read StationXML if available
        inv = read_inventory(xml_path) if xml_ok else None
        file_units = _get_input_units_from_inv(inv) if inv is not None else "unknown"
        if file_units:
            units_seen.add(file_units)
        if summary_units is None and file_units:
            summary_units = file_units
        
        # PPSD mean at 1/10/50/100 Hz (if inventory available)
        ppsd_vals = {"1Hz": None, "10Hz": None, "50Hz": None, "100Hz": None}
        asd_vals = {"1Hz": None, "10Hz": None, "50Hz": None, "100Hz": None}
        
        if inv is not None:
            try:
                # Per-file PPSD for summary values
                ppsd = PPSD(
                    tr.stats,
                    inv,
                    ppsd_length=60.0,
                    overlap=0.5,
                        special_handling=_special_handling,
                        db_bins=(-200, -20, 1.0),
                    period_limits=(0.01, 100)
                )
                ppsd.add(st)
                periods, mean = ppsd.get_mean()
                freqs = 1.0 / periods
                key_freqs = [1, 10, 50, 100]
                for kf in key_freqs:
                    if kf <= freqs.max() and kf >= freqs.min():
                        idx = int(np.argmin(np.abs(freqs - kf)))
                        ppsd_vals[f"{kf}Hz"] = float(mean[idx])
                        asd_vals[f"{kf}Hz"] = float(mean[idx] / 2.0)
                
                # Combined PPSD for summary plot
                if combined_ppsd is None:
                    combined_ppsd = PPSD(
                        tr.stats,
                        inv,
                        ppsd_length=60.0,
                        overlap=0.5,
                        special_handling=_special_handling,
                        db_bins=(-200, -20, 1.0),
                        period_limits=(0.01, 100)
                    )
                combined_ppsd.add(st)
            except Exception:
                pass
        
        summary_rows.append({
            "file": base,
            "xml": "yes" if xml_ok else "no",
            "units": file_units,
            "ppsd_1Hz": ppsd_vals["1Hz"],
            "ppsd_10Hz": ppsd_vals["10Hz"],
            "ppsd_50Hz": ppsd_vals["50Hz"],
            "ppsd_100Hz": ppsd_vals["100Hz"],
            "asd_1Hz": asd_vals["1Hz"],
            "asd_10Hz": asd_vals["10Hz"],
            "asd_50Hz": asd_vals["50Hz"],
            "asd_100Hz": asd_vals["100Hz"],
        })
    
    # Print summary table
    units_label = summary_units if summary_units else "unknown"
    if len(units_seen) > 1:
        units_label = "mixed"
    print(
        "File | XML | Units | PPSD(1/10/50/100 Hz) dB re 1 units^2/Hz | "
        "ASD(1/10/50/100 Hz) dB re 1 units/√Hz"
    )
    print("-" * 115)
    for row in summary_rows:
        ppsd_list = [row["ppsd_1Hz"], row["ppsd_10Hz"], row["ppsd_50Hz"], row["ppsd_100Hz"]]
        asd_list = [row["asd_1Hz"], row["asd_10Hz"], row["asd_50Hz"], row["asd_100Hz"]]
        ppsd_str = ", ".join(["{:.1f}".format(v) if v is not None else "n/a" for v in ppsd_list])
        asd_str = ", ".join(["{:.1f}".format(v) if v is not None else "n/a" for v in asd_list])
        print(f"{row['file']} | {row['xml']} | {row['units']} | {ppsd_str} | {asd_str}")
    
    # Summary plot: Combined PPSD over all files with XML
    if combined_ppsd is not None:
        print("\nSummary plot (combined PPSD):")
        fig = combined_ppsd.plot(xaxis_frequency=True, show=False)
        unit_note = "mixed units" if units_label == "mixed" else units_label
        if fig and fig.axes:
            fig.axes[0].set_ylabel(f"Power Spectral Density (dB re 1 {unit_note}^2/Hz)")
            fig.axes[0].set_xlabel("Frequency (Hz)")
        fig.suptitle(
            f"Combined PPSD (power, frequency): {len(summary_rows)} files ({unit_note})",
            y=1.02,
        )
        plt.show()
else:
    print("No files selected. Run Cell 4 first.")
```
